```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv('/content/Airline.csv')
df.head()
```

```python
df.info()
```

```python
# Missing value check
df.isna().sum()
```

```python
# Duplicate row check
df.duplicated().sum()
```

```python
# Cardinality check
df.nunique().sort_values()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Age"], kde=True, bins=30, color="steelblue")
plt.title("Age Distribution (Histogram + KDE)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["Gender"].value_counts().plot(kind="bar", color="teal")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Flight Status"].value_counts().plot(kind="bar", color="navy")
plt.title("Flight Status Distribution")
plt.xlabel("Flight Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df["Nationality"].value_counts().head(20).plot(kind="bar", color="purple")
plt.title("Top 20 Nationalities")
plt.xlabel("Nationality")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
